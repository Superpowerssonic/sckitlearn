```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### 1.  데이터를 읽어와서 결혼식 참석 인원(total), 식수 인원(reception)을 각각 변수로 저장

```python
dataset = pd.read_csv('QuizData.csv')
dataset[:5]
```

```python
X = dataset.iloc[:,:-1].values # 결혼식 참석인원
y = dataset.iloc[:,-1].values # 식수인원 
X
```

### 2. 전체 데이터를 훈련 세트와 테스트 세트로 분리. 비율은 75:25

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

### 3. 훈련 세트를 이용하여 단순 선형 회귀 모델 생성

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train,y_train)
```

### 4. 데이터 시각화(훈련 세트) 코드 작성

```python
plt.scatter(X_train, y_train, color='blue')
plt.plot(X_train, reg.predict(X_train), color='green')
plt.title('Wedding reception (train)')
plt.xlabel('totla')
plt.ylabel('reception')
plt.show()
```

```python
plt.scatter(X,y,color='blue') #산점도
plt.plot(X,reg.predict(X),color='green') # X에 대한 선형회귀 값이 Y 값으로 선 그래프로 그려짐
plt.title('Score by hours(genius)')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

### 5. 데이터 시각화(테스트 세트)

```python
plt.scatter(X_test, y_test, color='blue')
plt.plot(X_train, reg.predict(X_train), color='red')
plt.plot(X_test, reg.predict(X_test), color='green')
plt.title('Wedding reception (test)')
plt.xlabel('totla')
plt.ylabel('reception')
plt.xlim(75,400)
plt.ylim(50,235)
plt.show()
```

### 6. 훈련 세트, 테스트 세트에 대해 각각 모델 평가 점수 구하시오

```python
reg.score(X_train, y_train) # 훈련세트 평가 점수
reg.score(X_test, y_test) # 테스트 세트 평가 점수
```

### 7. 결혼식 참석 인원이 300명일 때 예상되는 식수 인원 구하시오

```python
total = 300
y_pred = reg.predict([[total]])
print(f'결혼식 참석 인원 {total}명에 대한 예상 식수 인원은 {np.around(y_pred[0]).astype(int)}입니다')
```
